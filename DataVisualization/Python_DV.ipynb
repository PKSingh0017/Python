# Data Visualization

It is a very important and necessary topic which is useful in understanding the data, extracting information and gathering patterns. One of the fundamental part of a Data Scientist's toolkit is data visualization. It's easy to create one but difficult to create a good one. <br>
 
There two primary uses of it: <br>
- To <i>Explore</i> data
- To <i>Communicate</i> data

There are many libraries that support data visualization like matplotlib, seaborn, Bokeh(similar to D3.js a JavaScript library for producing sophisticated interactive visualizations for the web), ggplot, pygal, plotly,geoplotlib, Gleam etc. The one which is used widely for many purposes is the matplotlib library, which we are going to learn here.

## Matplotlib

We will be using the <b>matplotlib.pyplot</b> module. pyplot maintains an internal state in which we build up a visualization step by step. Once all the steps are done we can save or show it.

```python
import matplotlib.pyplot as plt
```

```python
semesters = [1,2,3,4,5,6,7,8]
grades = [8.5,7.9,8.3,7.5,9.0,8.8,8.5,8.8]
```

```python
plt.title("Student Performance")
plt.xlabel("Semester")
plt.ylabel("Grade")
plt.plot(semesters, grades, color='violet', marker='o', linestyle='solid')
plt.show() # Displays the plot
```

### Bar Charts

Lets use the same data and plot that using bar charts

```python
plt.bar(semesters, grades, color ='indigo')
plt.ylabel("Grade")
plt.xlabel("Semester")
plt.title("Performance of a student")
plt.show()
```

A bar chart can be a useful choice for plotting histograms of grouped numeric values to visualize how the values are distributed

```python
from collections import Counter
```

```python
# Lets take grades of 15 students of a class
grades = [8.1,6.8,9.2,7.8,8.8,8.4,8.6,9.3,9.5,6.5,6.2,7.6,8.2,8.0,8.8]
groups = lambda grade: grade // 1 
histogram = Counter(groups(grade) for grade in grades)

plt.bar(histogram.keys(), histogram.values(), color='blue')
plt.xlabel("Distribution of grades")
plt.ylabel("Number of students")
plt.title("Distribution of Grades")
plt.show()
```

### Line Charts

By default plt.plot() uses this one. These are good for showing trends over a timeflow.

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip( variance, bias_squared) ]
xs = [i for i, _ in enumerate( variance) ]
```

When we call plt.plot multiple times all those calls merge and show in the same graph

```python
plt.plot(xs, variance, 'g-', label='variance') # Here 'g-' means green line plots
plt.plot(xs, bias_squared,'b-' , label='bias sqared') 
plt.plot(xs, total_error, 'r:', label='total error') # 'r:' means red dotted plots
plt.legend(loc=9) # It will display the names of the plots (loc=9 means at the center)
plt.xlabel("Model Complexity")
plt.title("Bias-Variance TradeOff")
plt.show()
```

### Scatterpots

Scatterplots are good for visualizing the relationship between two paired sets of data.

```python
age = [x*5 for x in range(15)]
power_level = [200, 500, 900, 1500, 1800, 2000,2500,3000, 4000, 4500, 4000, 2800, 2000, 1000, 900]
```

```python
plt.scatter(age, power_level, color='red')
plt.xlabel("age")
plt.ylabel("power level")
```

Some times when we are scattering comparable varibles we may have some problem with the scales

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
```

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title("Not on same scale")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title("On same scale")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.axis("equal")
plt.show()
```
